```python
import graphviz
def gv(s): return graphviz.Source('digraph G{ rankdir="LR"'+s + '; }');
```

```python
gv('''
splines=line
subgraph cluster_1{
    style=filled;
    color=lightgrey;
    "X1"
    "X2"
    ".."
    "X784"
    label = "Layer: "
}
subgraph cluster_2{
    style=filled;
    color=lightgrey;
    "X1" -> "node1"
    "X2" -> "node1"
    ".." -> "node1"
    "X784" -> "node1"
    
    "X1" -> "node2"
    "X2" -> "node2"
    ".." -> "node2"
    "X784" -> "node2"
    
    "X1" -> "..."
    "X2" -> "..."
    ".." -> "..."
    "X784" -> "..."
    
    "X1" -> "node32"
    "X2" -> "node32"
    ".." -> "node32"
    "X784" -> "node32"

    label = "Layer: ReLU"
}
subgraph cluster_3{
    style=filled;
    color=lightgrey;
    "node1" -> "y1"
    "node2" -> "y1"
    "..." -> "y1"
    "node32" -> "y1"
    
    "node1" -> "y2"
    "node2" -> "y2"
    "..." -> "y2"
    "node32" -> "y2"    
    label = "Layer: Softmax"
}
''')
```
