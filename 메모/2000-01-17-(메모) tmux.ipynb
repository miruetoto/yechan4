`tmux`는 터미널 멀티플렉서로, 하나의 터미널 세션 내에서 여러 창을 열고, 각 창에서 여러 창 패널을 생성할 수 있게 해줍니다. 이는 특히 원격 서버에 접속하여 여러 작업을 병렬로 수행하거나, 세션을 종료해도 작업을 지속할 수 있는 상황에서 매우 유용합니다.

아래는 `tmux`의 기본적인 사용 방법입니다.

### 1. `tmux` 시작하기
```bash
tmux
```
이 명령어는 새 `tmux` 세션을 시작합니다.

### 2. 새로운 세션 생성하기
새로운 `tmux` 세션을 시작하면서 세션 이름을 지정할 수 있습니다:
```bash
tmux new -s 세션이름
```

### 3. `tmux` 세션 목록 보기
현재 실행 중인 `tmux` 세션을 보려면:
```bash
tmux ls
```

### 4. `tmux` 세션에 재접속하기
특정 세션에 재접속하려면:
```bash
tmux attach -t 세션이름
```

### 5. `tmux` 세션에서 나가기
세션을 종료하지 않고 `tmux`에서 나가려면:
```bash
Ctrl+b, d
```
위의 키 조합을 사용하면 세션은 백그라운드에서 계속 실행됩니다.

### 6. 창 관리
- 새 창 생성: `Ctrl+b, c`
- 창 사이 이동: `Ctrl+b`, 숫자 (예: `Ctrl+b, 1`)
- 창 닫기: `exit` 또는 `Ctrl+d`
- 창 목록 보기: `Ctrl+b, w`

### 7. 창 패널 관리
- 패널 수직 분할: `Ctrl+b, %`
- 패널 수평 분할: `Ctrl+b, "`
- 패널 간 전환: `Ctrl+b, 화살표키` (또는 `Ctrl+b, o`)
- 패널 닫기: `exit` 또는 `Ctrl+d`
- 패널 크기 조절: `Ctrl+b`, 화살표키 (Ctrl을 누른 상태에서 화살표키로 조절)

### 8. `tmux` 세션 종료
세션을 완전히 종료하려면 모든 창과 패널을 닫거나, 다음 명령어를 입력합니다:
```bash
exit
```

### 9. `tmux` 설정
기본적으로 `~/.tmux.conf` 파일을 만들어 `tmux`의 기본 동작을 커스터마이징할 수 있습니다. 예를 들어, `tmux`에서 사용되는 기본 키 바인딩을 변경하려면 이 파일에 관련 설정을 추가하면 됩니다.

### 10. 자주 사용하는 명령어 요약
- `Ctrl+b, ?`: `tmux`의 모든 단축키 보기
- `Ctrl+b, [`: 복사 모드로 진입 (텍스트 스크롤 및 복사 가능)
- `Ctrl+b, d`: 세션 분리 (detach)
- `Ctrl+b, :`: 명령어 모드로 진입 (예: 세션 이름 변경 등)

이러한 기본적인 사용법을 익히고 나면 `tmux`를 사용해 보다 효율적으로 터미널 작업을 수행할 수 있습니다. 추가적으로 `tmux` 플러그인 매니저(TPM)를 사용하여 다양한 플러그인을 설치하고 관리할 수도 있습니다.
