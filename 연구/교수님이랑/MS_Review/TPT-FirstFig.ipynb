```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages
plt.rcParams.update({
    "text.usetex": True,
#    "font.family": "Helvetica"
})
# Generate data
np.random.seed(0)
n = 1000
x = np.arange(0, n)
y = np.random.normal(0, 1, n)

# Creating a sine wave for the second half
y[n//2:] = np.sin(2 * np.pi * (x[n//2:] - n//2) / (n // 2))

# Plot data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# First subplot (conventional plot)
ax1.plot(x, y, color='black', solid_capstyle='round')
ax1.set_title("Conventional Plot")

# Second subplot (varying thickness)
ax2.plot(x[:n//2], y[:n//2], color='black', solid_capstyle='round')
ax2.plot(x[n//2:], y[n//2:], color='black', linewidth=15, solid_capstyle='round')
ax2.set_title("Unconventional Plot with Varying Thickness")

# Save the figure
plt.savefig('./Dropbox/03_Yechan3/연구/교수님이랑/MS_Review/tpt_firstfig.pdf')

# Display plot
plt.show()
```
