```python
import numpy as np
import matplotlib.pyplot as plt 
plt.rcParams['text.usetex'] = True
```

```python
# Define the time vector
t = np.linspace(1, 1000, 2000) / 2000

# Compute the signal components
y1 = np.cos(62 * np.pi * t)
y2 = np.cos(58 * np.pi * t)
am = 2 * np.cos(2 * np.pi * t)
y = y1 + y2
```

```python
# Create the subplots with captions
fig, axs = plt.subplots(3, 1, figsize=(8, 5), tight_layout=True)

# First subplot
axs[0].plot(t, y1, color='C0')
axs[0].set_xlim(0, 0.5)
axs[0].set_ylim(-2, 2)
axs[0].set_title(r"Signal $\cos(62\pi t_n)$", fontsize=12)

# Second subplot
axs[1].plot(t, y2, color='C1')
axs[1].set_xlim(0, 0.5)
axs[1].set_ylim(-2, 2)
axs[1].set_title(r'Signal $\cos(58\pi t_n)$', fontsize=12)

# Third subplot
axs[2].plot(t, y1 + y2, color='black')
axs[2].plot(t, y1, color='C0', alpha=0.5, linewidth=0.5)
axs[2].plot(t, y2, color='C1', alpha=0.5, linewidth=0.5)
axs[2].plot(t, am, 'k--')
axs[2].set_xlim(0, 0.5)
axs[2].set_ylim(-2, 2)
axs[2].set_title(r'Combined Signal $\cos(62\pi t_n)+\cos(58\pi t_n) = 2\cos(2\pi t_n)\cos(60 \pi t)$', fontsize=12)
#plt.savefig('./Dropbox/03_Yechan3/연구/교수님이랑/MS_Review/beat.pdf')
plt.savefig('./beat.pdf')
```
