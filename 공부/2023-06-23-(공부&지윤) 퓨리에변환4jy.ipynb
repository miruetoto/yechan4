`-` 관측한자료 

```python
N=1000
x=np.linspace(0,1,N)
eps = np.random.randn(N)
X0 = np.sin(x*0*np.pi)
X1 = np.sin(x*2*np.pi)
X2 = np.sin(x*4*np.pi)
X3 = np.sin(x*6*np.pi)

y=2*X1+1*X2+3*X3+eps
```

```python
plt.plot(x,y,'o',alpha=0.1)
```

- observed signal

`-` 위의 자료를 해석하는 방법 

```python
def spec(y):
    N= len(y)
    return abs(np.fft.fft(y)/N)*2 
```

```python
y=2*X1+1*X2+3*X3+eps
yfft =spec(y) 
y1=2*X1
y2=1*X2
y3=3*X3
yfft1=spec(y1)
yfft2=spec(y2)
yfft3=spec(y3)
epsfft=spec(eps)
```

```python
plt.plot(yfft[:20],'o',alpha=0.5)
plt.plot(yfft1[:20],'x',alpha=1,)
plt.plot(yfft2[:20],'x',alpha=1)
plt.plot(yfft3[:20],'x',alpha=1)
plt.plot(epsfft[:20],'x',alpha=1)
```
